```python
# Final Compression Script

def compress(text):
    if not text:
        return ""
    compressed = ""
    count = 1
    for i in range(1, len(text)):
        if text[i] == text[i - 1]:
            count += 1
        else:
            compressed += text[i - 1] + str(count)
            count = 1
    compressed += text[-1] + str(count)  # add last character
    return compressed

# Ensure input.txt exists
import os
if not os.path.exists("input.txt"):
    with open("input.txt", "w") as f:
        f.write("aaabbbbcc")  # default sample text

# Read input
with open("input.txt", "r") as f:
    data = f.read().strip()

# Compress
compressed_data = compress(data)

# Write compressed output
with open("compressed.txt", "w") as f:
    f.write(compressed_data)

# Print results
print("Original:", data)
print("Compressed:", compressed_data)
```
